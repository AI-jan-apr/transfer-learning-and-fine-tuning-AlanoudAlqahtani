```python
import os
import glob
import json
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

print("TensorFlow version:", tf.__version__)
print("Available GPUs:", tf.config.list_physical_devices('GPU'))

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)
```

```python
def get_data_extract():
    if "dataset" in os.listdir():
        print("Dataset already exists")
    else:
        print("Downloading the data...")
        !wget -O food-data.zip https://www.kaggle.com/api/v1/datasets/download/trolukovich/food11-image-dataset

        print("Extracting data...")
        os.makedirs("dataset", exist_ok=True)
        !unzip -q food-data.zip -d dataset
        print("Done.")

get_data_extract()
```

```python
all_paths = glob.glob("dataset/*/*/*.jpg")
print("Total images found:", len(all_paths))
print("Sample path:", all_paths[0] if len(all_paths) > 0 else "No images found")
```

```python
labels_str = [p.split("/")[-2] for p in all_paths]
class_names = sorted(list(set(labels_str)))
num_classes = len(class_names)

print("Classes:", class_names)
print("Number of classes:", num_classes)
```

```python
df = pd.DataFrame({
    "filepath": all_paths,
    "label_str": labels_str
})

class_to_index = {name: idx for idx, name in enumerate(class_names)}
index_to_class = {idx: name for name, idx in class_to_index.items()}

df["label"] = df["label_str"].map(class_to_index)
df = df.sample(frac=1, random_state=SEED).reset_index(drop=True)

df.head()
```

```python
train_df, temp_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df["label"],
    random_state=SEED
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df["label"],
    random_state=SEED
)

print("Train size:", len(train_df))
print("Validation size:", len(val_df))
print("Test size:", len(test_df))
```

```python
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

train_aug = keras.Sequential([
    layers.Resizing(IMG_SIZE[0], IMG_SIZE[1]),
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
], name="train_augmentation")

val_aug = keras.Sequential([
    layers.Resizing(IMG_SIZE[0], IMG_SIZE[1]),
], name="val_augmentation")
```

`tf.data` builds a performance model of the input pipeline and runs an optimization algorithm to find a good allocation of its CPU budget across all parameters specified as `AUTOTUNE`. While the input pipeline is running, `tf.data` tracks the time spent in each operation, so that these times can be fed into the optimization algorithm.

```python
def load_image(path, label, augmenter):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.cast(image, tf.float32)
    image = augmenter(image)
    return image, label

def make_dataset(dataframe, augmenter, training=False):
    ds = tf.data.Dataset.from_tensor_slices(
        (dataframe["filepath"].values, dataframe["label"].values)
    )

    if training:
        ds = ds.shuffle(len(dataframe), seed=SEED)

    ds = ds.map(
        lambda x, y: load_image(x, y, augmenter),
        num_parallel_calls=AUTOTUNE
    )
    ds = ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return ds

train_ds = make_dataset(train_df, train_aug, training=True)
val_ds = make_dataset(val_df, val_aug, training=False)
test_ds = make_dataset(test_df, val_aug, training=False)
```

```python
images, labels_batch = next(iter(train_ds))

plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(index_to_class[int(labels_batch[i])])
    plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
def build_efficientnet_model(num_classes, trainable=False, dropout_rate=0.3):
    base_model = keras.applications.EfficientNetB0(
        include_top=False,
        weights="imagenet",
        input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3)
    )

    base_model.trainable = trainable

    inputs = keras.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
    x = base_model(inputs, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(dropout_rate)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = keras.Model(inputs, outputs)
    return model, base_model
```

```python
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor="val_loss",
        patience=4,
        restore_best_weights=True
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.2,
        patience=2,
        min_lr=1e-7
    )
]
```

```python
feature_model, feature_base = build_efficientnet_model(
    num_classes=num_classes,
    trainable=False,
    dropout_rate=0.3
)

feature_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_feature = feature_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=15,
    callbacks=callbacks
)
```

```python
def plot_history(history, title):
    acc = history.history["accuracy"]
    val_acc = history.history["val_accuracy"]
    loss = history.history["loss"]
    val_loss = history.history["val_loss"]
    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, label="Train Accuracy")
    plt.plot(epochs, val_acc, label="Validation Accuracy")
    plt.title(f"{title} - Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, label="Train Loss")
    plt.plot(epochs, val_loss, label="Validation Loss")
    plt.title(f"{title} - Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_history(history_feature, "Feature Extraction")
```

```python
feature_test_loss, feature_test_acc = feature_model.evaluate(test_ds, verbose=0)

print("Feature Extraction Test Loss:", feature_test_loss)
print("Feature Extraction Test Accuracy:", feature_test_acc)
```

```python
def get_predictions(model, dataset):
    y_true = []
    y_pred = []

    for x_batch, y_batch in dataset:
        preds = model.predict(x_batch, verbose=0)
        pred_labels = np.argmax(preds, axis=1)

        y_true.extend(y_batch.numpy())
        y_pred.extend(pred_labels)

    return np.array(y_true), np.array(y_pred)

def plot_confusion_matrix(y_true, y_pred, class_names, title):
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    plt.imshow(cm, cmap="Blues")
    plt.title(title)
    plt.colorbar()
    plt.xticks(range(len(class_names)), class_names, rotation=90)
    plt.yticks(range(len(class_names)), class_names)

    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, cm[i, j], ha="center", va="center")

    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()
```

```python
y_true_feat, y_pred_feat = get_predictions(feature_model, test_ds)

print("Classification Report - Feature Extraction")
print(classification_report(y_true_feat, y_pred_feat, target_names=class_names))

plot_confusion_matrix(
    y_true_feat,
    y_pred_feat,
    class_names,
    "Confusion Matrix - Feature Extraction"
)
```

fine tuning

```python
def unfreeze_last_n_layers(base_model, n=30):
    base_model.trainable = True

    for layer in base_model.layers[:-n]:
        layer.trainable = False

    for layer in base_model.layers:
        if isinstance(layer, layers.BatchNormalization):
            layer.trainable = False
```

```python
fine_tune_model = feature_model
base_model = fine_tune_model.layers[1]

unfreeze_last_n_layers(base_model, n=30)

fine_tune_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_finetune = fine_tune_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=callbacks
)
```

```python
plot_history(history_finetune, "Fine-Tuning")
```

```python
fine_tune_test_loss, fine_tune_test_acc = fine_tune_model.evaluate(test_ds, verbose=0)

print("Fine-Tuning Test Loss:", fine_tune_test_loss)
print("Fine-Tuning Test Accuracy:", fine_tune_test_acc)
```

```python
y_true_ft, y_pred_ft = get_predictions(fine_tune_model, test_ds)

print("Classification Report - Fine-Tuning")
print(classification_report(y_true_ft, y_pred_ft, target_names=class_names))

plot_confusion_matrix(
    y_true_ft,
    y_pred_ft,
    class_names,
    "Confusion Matrix - Fine-Tuning"
)
```
